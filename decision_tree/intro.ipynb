
## What is entropy ?

https://en.wikipedia.org/wiki/Entropy_(information_theory)

- Information entropy is the average amount of information conveyed by an event, when considering all possible outcomes. 
- Entropy is the expected value of self-information of a variable.


$$ {\displaystyle \mathrm {H}(X)=-\sum _{i=1}^{n}{\mathrm {P} (x_{i})\log \mathrm {P} (x_{i})}} $$




### Introduction 

The more surprising the outcome (low probability), the higher the informational value of a message. The information content/surprisal/self-information of an event $E$ is defined as follows :


$$ I(E) = log_{2}(\frac{1}{P(E)}) $$


which gives ${\displaystyle -\log(1)=0 }$ surprise when the probability of the event is 1 and infinite surpise ${\displaystyle -\log(0)=\infty }$

- Note that, $log$ is used as it is only function that satisfies this specific set of characterization. There are some axioms for which it needs to fulfil, but it is out of scope. Just remember that $log$ function has the properties we desire to define information value of a random variable.

```python
# Generate data
x = np.random.uniform(low=0.0, high=10, size=1000)
y = np.log2(x)

fig, ax = plt.subplots(1,2, figsize=(15,6))

p = sns.lineplot(x=x,y=y,ax=ax[0])
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=1, color='r')
p = ax[0].set(title='log(x) vs x',xlabel = "x", ylabel = "$log_{2}(x)$")

x = np.random.uniform(low=0.0, high=1.0, size=100)
y = -np.log2(x)

p = sns.lineplot(x=x,y=y,ax=ax[1])
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=1, color='r')
p = ax[1].set(title='-log(x) vs x',xlabel = "x", ylabel = "$-log_{2}(x)$")
```
